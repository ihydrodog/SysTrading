```python
import pandas as pd
import pandas_datareader as web
import datetime
import numpy as np


def download_dataframe( file_name, company_code, year1, month1, date1, year2, month2, date2 ):
    start = datetime.datetime( year1, month1, date1 )
    end = datetime.datetime( year2, month2, date2 )

    df = web.DataReader( "%s.KS" % (company_code), "yahoo", start, end )
    df.to_pickle( file_name )

    return df
```

```python
df = download_dataframe( 'samsung', '005930', 2018,1,1,2019,1,1 )
```
